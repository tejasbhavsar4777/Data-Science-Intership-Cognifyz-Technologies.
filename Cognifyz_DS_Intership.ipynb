```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Encode 'City'

df['Restaurant Name_encoded'] = label_encoder.fit_transform(df['Restaurant Name'])

df['Switch to order menu_encoded'] = label_encoder.fit_transform(df['Switch to order menu'])



df['Currency'] = label_encoder.fit_transform(df['Currency'])


#df['Is delivering now'] = label_encoder.fit_transform(df['Is delivering now'])


df['City_encoded'] = label_encoder.fit_transform(df['City'])

# Encode 'Rating'
df['Address_encoded'] = label_encoder.fit_transform(df['Address'])

df['Locality_encoded'] = label_encoder.fit_transform(df['Locality'])

df['Locality_Verbose_encoded'] = label_encoder.fit_transform(df['Locality Verbose'])

df['Cuisines_encdode'] = label_encoder.fit_transform(df['Cuisines'])


df['Rating_text_encoded'] = label_encoder.fit_transform(df['Rating text'])


df['Rating_color_encoded'] = label_encoder.fit_transform(df['Rating color'])


print("\nAfter Label Encoding:")

```

```python
X=df.drop(['Aggregate rating'],axis=1)
y=df['Aggregate rating']
```

#### Experiment with different algorithms (e.g., linear regression, decision trees, random forest) and compare their performance.

```python
#Experiment with different algorithms (e.g., linear regression, decision trees, random forest) and compare their performance.
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42, n_estimators=100)
}
```

```python
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results[name] = {
        'Mean Squared Error': mse,
        'R-squared': r2
    }

for model, metrics in results.items():
    print(f"Model: {model}")
    for metric, value in metrics.items():
        print(f"  {metric}: {value}")
    print()
```
